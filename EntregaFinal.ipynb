```python
# Vamos a crear una función que reciba el año de estreno y nos devuelva una etiqueta

def categorizar_estreno(año):
    """
    Función para categorizar años de estreno.

    Parámetros:
    - año (int): Número que se va a categorizar.

    Retorna:
    - str: Categoría a la que pertenece el año.
    """
    if año <= 1999:
        return 'Película antigua'
    elif 2000 <= año <= 2009:
        return 'Película de hace 20 años'
    elif 2010 <= año <= 2019:
        return 'Película de hace 10 años'
    else:
        return 'Película nueva'
```

```python
# Ahora que tenemos la función creada vamos a aplicarla para nuestro DataFrame y para la columna del año de estreno
# Nota: Vamos a usar el total_2 para no realizar cambios importantes sobre el DataFrame original

df_total_2 = df_total
df_total_2["year category"] = df_total_2["release_year"].apply(categorizar_estreno)
df_total_2.head(5)
```

```python
# Con apply también podemos cambiar el tipo de dato de la columna. Vamos a ver qué tipo de dato tiene la columna del año de estreno

df_total_2["release_year"].dtype
```

```python
# Vamos a cambiarla a tipo float

# Definimos la función:

def cambiar_float(año):
    """
    Función para cambiar el tipo de un int a float.

    Parámetros:
    - año (int): Número que se va a convertir.

    Retorna:
    - float: El número proporcionado con el tipo cambiado a float.
    """
    return float(año)
```

```python
# Llamamos a la función

df_total_2["release_year"] = df_total_2["release_year"].apply(cambiar_float)
```

```python
# Comprobamos

df_total_2["release_year"].dtype
```

```python
# lambda

# Vamos a hacer lo mismo para la columna de duración, volviéndola a pasar a int

df_total_2["release_year"] = df_total_2["release_year"].apply(lambda x: int(x))
```

```python
# Comprobamos

df_total_2["release_year"].dtype
```

```python
# Lambda con condicionales

df_total_2["type"] = df_total_2["type"].apply(lambda x: "Película" if x == "Movie" else "Serie")
df_total_2.head(5)
```

```python
# Podemos usar una función lambda para utilizar funciones a las que le pasamos más de un parámetro

# Primero definiremos la función:

def serie_2015(col_serie, col_2015):
    """
    Función para identificar un usuario con unas características concretas.

    Parámetros:
    - col_serie (str): Columna que marca si el contenido es una serie
    - col_2015 (str): Columna que marca si el año de estreno es el 2015

    Retorna:
    - bool: Si las condiciones se cumplen o no
    """
    if col_serie == "TV Show" and col_2015 == "2015":
        return True
    else:
        return False
```

```python
# Ahora llamamos a la función utilizando apply y lambda

df_total_2["serie 2015"] = df_total_2.apply(lambda x: serie_2015(x["type"], x["release_year"]), axis=1)
df_total_2.head(5)
```
