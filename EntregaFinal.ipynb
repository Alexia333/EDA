```python
# Vamos a crear una funci√≥n que reciba el a√±o de estreno y nos devuelva una etiqueta

def categorizar_estreno(a√±o):
    """
    Funci√≥n para categorizar a√±os de estreno.

    Par√°metros:
    - a√±o (int): N√∫mero que se va a categorizar.

    Retorna:
    - str: Categor√≠a a la que pertenece el a√±o.
    """
    if a√±o <= 1999:
        return 'Pel√≠cula antigua'
    elif 2000 <= a√±o <= 2009:
        return 'Pel√≠cula de hace 20 a√±os'
    elif 2010 <= a√±o <= 2019:
        return 'Pel√≠cula de hace 10 a√±os'
    else:
        return 'Pel√≠cula nueva'
```

```python
# Ahora que tenemos la funci√≥n creada vamos a aplicarla para nuestro DataFrame y para la columna del a√±o de estreno
# Nota: Vamos a usar el total_2 para no realizar cambios importantes sobre el DataFrame original

df_total_2 = df_total
df_total_2["year category"] = df_total_2["release_year"].apply(categorizar_estreno)
df_total_2.head(5)
```

```python
# Con apply tambi√©n podemos cambiar el tipo de dato de la columna. Vamos a ver qu√© tipo de dato tiene la columna del a√±o de estreno

df_total_2["release_year"].dtype
```

```python
# Vamos a cambiarla a tipo float

# Definimos la funci√≥n:

def cambiar_float(a√±o):
    """
    Funci√≥n para cambiar el tipo de un int a float.

    Par√°metros:
    - a√±o (int): N√∫mero que se va a convertir.

    Retorna:
    - float: El n√∫mero proporcionado con el tipo cambiado a float.
    """
    return float(a√±o)
```

```python
# Llamamos a la funci√≥n

df_total_2["release_year"] = df_total_2["release_year"].apply(cambiar_float)
```

```python
# Comprobamos

df_total_2["release_year"].dtype
```

```python
# lambda

# Vamos a hacer lo mismo para la columna de duraci√≥n, volvi√©ndola a pasar a int

df_total_2["release_year"] = df_total_2["release_year"].apply(lambda x: int(x))
```

```python
# Comprobamos

df_total_2["release_year"].dtype
```

```python
# Lambda con condicionales

df_total_2["type"] = df_total_2["type"].apply(lambda x: "Pel√≠cula" if x == "Movie" else "Serie")
df_total_2.head(5)
```

```python
# Podemos usar una funci√≥n lambda para utilizar funciones a las que le pasamos m√°s de un par√°metro

# Primero definiremos la funci√≥n:

def serie_2015(col_serie, col_2015):
    """
    Funci√≥n para identificar un usuario con unas caracter√≠sticas concretas.

    Par√°metros:
    - col_serie (str): Columna que marca si el contenido es una serie
    - col_2015 (str): Columna que marca si el a√±o de estreno es el 2015

    Retorna:
    - bool: Si las condiciones se cumplen o no
    """
    if col_serie == "TV Show" and col_2015 == "2015":
        return True
    else:
        return False
```

```python
# Ahora llamamos a la funci√≥n utilizando apply y lambda

df_total_2["serie 2015"] = df_total_2.apply(lambda x: serie_2015(x["type"], x["release_year"]), axis=1)
df_total_2.head(5)
```
